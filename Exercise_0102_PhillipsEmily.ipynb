1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.
- 10 rows
- 50 columns
- 1 <= x <= 500

```python
import numpy as np
import random
import pandas as pd
```

```python
row, col = 10,50
rand_df = pd.DataFrame([[random.randint(1,500) for x in range(col)]for y in range(row)])
```

```python
rand_df
```

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
print('Sum of each row: {}'.format(rand_df.sum(axis=1)))
```

```python
#Sum of each column
print('Sum of each column: {}'.format(rand_df.iloc[:].sum()))
```

```python
#Sum of all entries
print('Sum of all entries: {}'.format(rand_df.values.sum()))
```

3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
print('Minimum: {}'.format(rand_df.values.min()))
print('Maximum: {}'.format(rand_df.values.max()))
print('Average: {}'.format(rand_df.values.mean()))
print('Standard Deviation: {}'.format(rand_df.values.std()))
```

4. Sort the DataFrame: 
- sort on column 2
- sort on row 2

```python
#Sort by column 2
rand_df.sort_values(by=[2])
```

```python
#Sort on row
rand_df.sort_values(by=[2],axis=1)
```
